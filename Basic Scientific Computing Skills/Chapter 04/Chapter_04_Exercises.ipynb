### Question 2
The following equation defines the relationship between energy (J) of a photo and its wavelength (m) where h is Plank’s constant ($6.626 \cdot 10^-34 Js$) and c is the speed of light in a vacuum ($2.998 \cdot 10^8 m/s$).

$$ E = \frac{hc}{\lambda} $$


a. Generate an array containing the wavelengths of visible light ($4.00 \times 10^-7 m$ to $8.00 \times 10^-7 m$) in $5.00 \times 10^-8 m$ increments.

b. Generate a second array containing the energy of each wavelength of light from part a.

````python
# Generate an array of wavelengths from 4.00e-7 m to 8.00e-7 m in 5.00e-8 m increments
# np.arange(start, stop, step), we have to add 5.00e-8 to stop to include the endpoint
wavelengths = np.arange(4.00e-7, 8.00e-7 + 5.00e-8, 5.00e-8)
wavelengths
````

````python
def photon_energy(wavelengths):
    h = 6.626e-34  # Planck's constant in Js
    c = 2.998e8    # Speed of light in m/s
    energies = h * c / wavelengths
    return energies

energy = photon_energy(wavelengths)
energy
````

We created a NumPy array called `wavelengths` using a start/stop/step pattern.  
We then defined a function that calculates energy using ($E = \frac{hc}{\lambda}$).  
Because NumPy performs functions on arrays, we can pass the whole `wavelengths` array into the function and it returns an array of energies, which we stored as `energies`.

### Question 4
The following array contains temperatures in Fahrenheit. Convert these values to 
C without using a for loop.
```python
F = array([0, 32, 100, 212, 451])
```

````python
# formula (32°F − 32) × 5/9 = 0°C
def fahrenheit_to_celsius(F):
    F = np.asarray(F)          
    return (F - 32) * 5 / 9

F = np.array([0, 32, 100, 212, 451])
C = fahrenheit_to_celsius(F)
print (C)
````

### Question 5
Generate two arrays containing the following sine functions from x = $0 \ to \ 10  \pi$

$$ y = \sin (x) $$
$$ y = \sin (1.1x + 0.5) $$

- a Plot these two sine waves on the same plot.

- b Add the two sine functions together and plot the result.

- c Explain why the signal in part b is smaller in one area and larger in another. 
> Hint: look at your plot for part a to see how the two original sine waves related to each other.

````python
x =  np.linspace(0, 10 * np.pi, 100)  # 100 points from 0 to 10π, np.linspace(Start, Stop, No. of spaces)

y_1 = np.sin(x)  # sine of x
y_2 = np.sin(1.1*x+0.5)  # sine of (x + 1.1 + 0.5)
````

````python
import matplotlib.pyplot as plt
plt.plot(x, y_1, label='sin(x)')
plt.plot(x, y_2, label='sin(1.1 + x + 0.5)')
plt.xlabel('x')
plt.ylabel('y')
plt.title('Sine Functions')
plt.legend()
plt.show()
````

````python
y_3 = y_1 + y_2
plt.plot(x, y_3, label='sin(x) + sin(1.1 + x + 0.5)')
plt.xlabel('x')
plt.ylabel('y')
plt.title('Sum of Sine Functions')
plt.legend()
plt.show() 
````

This plot is due to the phase of the two sine waves starting with having constructive interference but as their phases shift further apart they become destructive

### Question 6
The numerical relationship between $ \Delta G \degree $ and $k$ (equilibrium constant) is shown below. 

$$ \Delta G \degree = -RTln(k) $$

Plot $ \Delta G \degree $ versus $k$ at standard temperature and pressure for K values of 0.001 to 1000. 
> Use NumPy arrays and do not use any for loops.

````python
k = np.linspace (0.001, 1000, 100000)

def gibbs_free (k):
    R = 8.314
    T = 273 # K
    dG = -R*T*np.log(k)
    return dG

dG = gibbs_free(k)
````

````python
plt.plot(k, dG, label='$ \Delta G \degree $ vs. $k$')
plt.xlabel('x')
plt.ylabel('y')
plt.title('numerical relationship between $ \Delta G \degree $ and $k$')
plt.legend()
plt.show() 
````

### Question 7
The numerical relationship between k (rate constant) and Ea is shown below. 

$$ k = Ae ^{-Ea/RT} $$

Plot k versus Ea at standard temperature and pressure for activation energies of 1 to 20 kJ/mol. 

> Use NumPy arrays, do not use any for loops, and use A = 1. Watch your energy units carefully.

````python
Ea_kJ = np.linspace(1, 20, 200)

def rate(Ea_kJ):
    R = 8.314      # J/(mol*K)
    T = 273.15     # K
    Ea_J = Ea_kJ * 1000.0
    return np.exp(-Ea_J / (R * T))  # A = 1

k = rate(Ea_kJ)

plt.plot(Ea_kJ, k, label='k vs Ea')
plt.xlabel('Ea (kJ/mol)')
plt.ylabel('k (A=1)')
plt.title('k versus Ea at STP')
plt.legend()
plt.show()
````

### Question 9 
Generating an Combining arrays – Bohr hydrogen atom.

a. Create an array containing the principal quantum numbers (n) for the first eight orbits of a hydrogen atom (i.e., 1 -> 8).

b. Generate a second array containing the energy (J) of each orbit in part A for a Bohr model of a hydrogen atom using the equation below.

$$ E = -2.18 \times 10^-18 J \frac{1}{n^2} $$

c. Combine the two arrays from parts A and B into a new 8 x 2 array with the first column containing the principal quantum numbers and the second containing the energies.

````python
n = np.arange(1,9,1)

def bohr_energy(n):
    energy = -2.18 * 10**-18 * 1/(n**2)
    return energy

J = bohr_energy(n)
J
````

````python
combined = np.column_stack((n, J))  # Combines n and J into a 2D array with n as the first column and J as the second
combined
````
