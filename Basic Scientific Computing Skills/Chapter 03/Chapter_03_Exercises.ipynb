### Question 1

Visualize the relationship between pressure and volume for 1.00 mol of He(g) at 298 K in an expandable vessel as it increases from 1 L to 20 L. R = 0.08206 L路atm/mol路K. 

This will require you to generate values and perform the calculation using the equation below.
$$PV = nRT$$


```python
# P = nRT/V plot of P against V as V goes from 1 to 20L
import matplotlib.pyplot as plt

R = 0.08206  # L路atm/(mol路K)
n = 1.00     # mol
T = 298      # K
volumes = list(range(1, 21))

def ideal_gas_pressure(n, R, T, V):
    P = (n * R * T) / V
    return P

# Calculate pressures for each volume
pressures = [ideal_gas_pressure(n, R, T, V) for V in volumes]

# Plot pressure vs volume
# Use Matplotlib's "tab:" colors (Tableau palette) instead of single-letter colors like 'b':
# the tab: palette matches Matplotlib's default color cycle, gives clearer/more consistent colors,
# and provides more distinct options than the limited single-letter set.
plt.plot(volumes, pressures, 'tab:red')
plt.xlabel("Volume (L)")
plt.ylabel("Pressure (atm)")
plt.title("Pressure vs Volume for 1.00 mol He(g) at 298 K")
plt.grid(True)
print("pressures[0] =", pressures[0], " last =", pressures[-1])
plt.show()
```

### Question 3
The following functions are an example of the sandwich theorem which aids in determining limits of function $g(x)$ by knowing its range is between $f(x)$ and $h(x)$ in the relevant domain. Plot all three functions on the same axes to show that $f(x) > g(x) > h(x)$ for x of -50 to 50.
> Be sure to include a legend.

$f(x) = x^2 \qquad g(x) = x^2sin(x) \qquad h(x) = -x^2$

```python
x = list(range(-50, 51))

def f(x):
    return x**2

def g(x):
    import math
    return x**2 * math.sin(x)

def h(x):
    return -x**2

#list comprehensions to compute function values
f_values = [f(xi) for xi in x]
g_values = [g(xi) for xi in x]
h_values = [h(xi) for xi in x]

plt.plot(x, f_values, label='f(x) = x^2', color='tab:blue')
plt.plot(x, g_values, label='g(x) = x^2 sin(x)', color='tab:orange')
plt.plot(x, h_values, label='h(x) = -x^2', color='tab:green')
plt.xlabel("x values from -50 to 50")
plt.ylabel("Function values")
plt.title("Plot of f(x), g(x), and h(x) displaying sandwich theorem")
plt.legend(loc="upper center")
plt.grid(True)
plt.show()
```

### Question 4
Plot the concentration of A with respect to time for the following elementary step if $k = 0.12 M^-1s^-1$ using the appropriate integrated rate law.

$$2A \rightarrow P$$

```python
rate_constant = 0.12  # M^-1 s^-1
initial_concentration_A = 1.0  # M
time_points = list(range(0, 101))  # seconds

def concentration_A(t, k, A0):
    return A0 / (1 + 2 * k * A0 * t)
# Integrated rate law for the elementary reaction 2A  P (second-order in A):
#   -d[A]/dt = 2k[A]^2    1/[A] = 1/[A]0 + 2kt    [A](t) = [A]0 / (1 + 2k[A]0 t)

concentrations_A = [concentration_A(t, rate_constant, initial_concentration_A) for t in time_points]

plt.plot(time_points, concentrations_A, 'tab:purple')
plt.xlabel("Time (s)")
plt.ylabel("Concentration of A (M)")
plt.title("Concentration of A over Time for 2A  P with $k = 0.12 M^-1s^-1$")
plt.grid(True)
plt.show()
```

### Question 12
Generate a standing wave surface plot using the following equation and parameters: 
$L$ = 1, $n_x$ = 2, $n_y$ = 2

$$\psi(x,y) = (2/L)sin(n_x\pi x/L)sin(n_x\pi /L)$$

```python
L = 1
nx = 2
ny = 2

x = np.linspace(0, L, 100)
y = np.linspace(0, L, 100)
X, Y = np.meshgrid(x, y)

def wave(x,y, nx, ny):
    psi = (2/L) * np.sin(nx * np.pi * X / L) * np.sin(ny * np.pi * Y / L)
    return psi

Z = wave(X, Y, nx, ny)

fig = plt.figure(figsize=(8, 5))
ax = fig.add_subplot(111, projection='3d')
ax.plot_surface(X, Y, Z, cmap='viridis')
ax.set_xlabel('X')
ax.set_ylabel('Y')
ax.set_zlabel('(x,y)')
ax.set_title('Standing Wave Surface Plot for $n_x=2$, $n_y=2$')
plt.show()
```
